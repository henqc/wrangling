```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('./data/airbnb_hw.csv', low_memory=False)
df.head()
```

```python
df.info()
```

```python
#Price is listed as an object because it is including a comma seperator for numeric values above 999

# Replacing the commas with empty space
price = df['Price'].str.replace(',','') 
#Casting to a numeric value
price = pd.to_numeric(price,errors='coerce') 
price.unique() 

```

```python
# Checking how many values are missing which is 0!!
print(sum(price.isnull())) 
```

Question 2 Part 2

```python
df = pd.read_csv('./data/sharks.csv', low_memory=False)
df.head()
```

```python
df['Type'].value_counts()
```

```python
#A lot of the columns have similiar column meaning so we will consolidate them

# Create a temporary vector
temp = df['Type'] 
# Combining all incidents relating to water vehicles into one col
temp = temp.replace(['Sea Disaster', 'Boatomg','Boat','Boating'],'Watercraft') 
# All unclean data
temp = temp.replace(['Unverified','Under investigation','Invalid', 'Questionable','Unconfirmed'],np.nan) 
temp.value_counts()

```

```python
# Replace with cleaned version
df['Type'] = temp 
del temp 
df['Type'].value_counts()
```

Question 2 Part 3

```python
url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
df = pd.read_csv(url, low_memory=False)
df.head()
```

```python
release = df['WhetherDefendantWasReleasedPretrial']
print(release.unique())
print(release.value_counts())
```

```python
# Accoring to the codebook, 9's should be replaced with nan values as they are unclear

#Replacing all 9's with nan
release = release.replace(9,np.nan) 
#Replacing with cleaned data
df['WhetherDefendantWasReleasedPretrial'] = release 
del release
df['WhetherDefendantWasReleasedPretrial'].value_counts()
```

Question 2 Part 4

```python
sentence = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']
print(sentence.unique())
```

```python
#some values are blank which are missing values

#Cast to a numeric value so we can convert the blanks to NaN
sentence = pd.to_numeric(sentence,errors='coerce')
# Create a temp just to check how many null values there are
sentence_NA = sentence.isnull() 
print( np.sum(sentence_NA),'\n') 
# Understanding how missing sentence lengths relate to the type of sentence imposed.
print( pd.crosstab(sentence_NA, type), '\n') 
```

```python
# Length set to 0 where charges are dismissed
sentence = sentence.mask( type == 4, 0) 
# Type 9 is unclear so replace with nan
sentence = sentence.mask( type == 9, np.nan) 

sentence_NA = sentence.isnull() 
print( pd.crosstab(sentence_NA, type), '\n')
print( np.sum(sentence_NA),'\n') 

# Replace with cleaned version
df['ImposedSentenceAllChargeInContactEvent'] = sentence 
del sentence, type 

```
